```python
import pandas as pd
import numpy as np
train_data = pd.read_csv("used_car_train_20200313.csv",sep=" ")
test_data=pd.read_csv("used_car_testB_20200421.csv",sep=' ')
```

```python
train_data
```

```python
test_data
```

```python
test_data['notRepairedDamage'].replace('-','0.0',inplace=True)
test_data['notRepairedDamage']=test_data['notRepairedDamage'].astype('float64')
```

```python
train_data['notRepairedDamage'].replace('-','0.0',inplace=True)
train_data['notRepairedDamage']=train_data['notRepairedDamage'].astype('float64')
```

```python
train_data['notRepairedDamage'].value_counts()
```

```python
test_data['notRepairedDamage'].value_counts()
```

```python
train_data=train_data.fillna(train_data.median())
test_data=test_data.fillna(test_data.median())
```

```python
train_data['power'][train_data['power']>600]=600
train_data['power'].describe()
```

```python
features=train_data.columns.tolist()
features.remove('price')
features.remove('SaleID')
features
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scalar=MinMaxScaler()
x=min_max_scalar.fit_transform(train_data[features].values)
x_test=min_max_scalar.transform(test_data[features].values)
```

```python
y=train_data['price']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow import keras
```

```python
# 搭建网络模型
model_k=keras.Sequential([
    keras.layers.Dense(250,activation='relu',input_shape=[len(features)]),
    keras.layers.Dense(250,activation='relu'),
    keras.layers.Dense(250,activation='relu'),
    keras.layers.Dense(1)
])
```

```python
model_k.compile(loss='mean_absolute_error',optimizer='Adam')
model_k.fit(train_x,train_y,batch_size=1024,epochs=40)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print('训练集的MEA',mean_absolute_error(train_y,model_k.predict(train_x)))
print('验证集的MEA',mean_absolute_error(test_y,model_k.predict(test_x)))
```

```python
# 搭建网络模型
model=keras.Sequential([
    keras.layers.Dense(250,activation='relu',input_shape=[len(features)]),
    keras.layers.Dense(250,activation='relu'),
    keras.layers.Dense(250,activation='relu'),
    keras.layers.Dense(1)
])
model.compile(loss='mean_absolute_error',optimizer='Adam')
model.fit(x,y,batch_size=1024,epochs=40)
```

```python
predict_y=model.predict(x_test)
result=pd.DataFrame()
result['SaleID']=test_data['SaleID']
result['price']=predict_y
result['price'][result['price']<11]=11
result.to_csv('./ans_nn.csv',index=False)
```
