```python
import numpy as np
from scipy.linalg import svd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
image = Image.open('256.bmp')
```

```python
image_np = np.array(image)
```

```python
plt.imshow(image_np, cmap=plt.cm.gray)
```

```python
# 对图像矩阵进行奇异值分解
p, s, q = svd(image_np, full_matrices=False)
```

```python
def get_image_feature(s, k):
    s_temp = np.zeros(s.shape[0])
    s_temp[0:k] = s[0:k]
    s = s_temp * np.identity(s.shape[0])
    temp = np.dot(p, s)
    temp = np.dot(temp, q)
    plt.imshow(temp, cmap=plt.cm.gray)
```

```python
image_np.shape[0]*image_np.shape[1]
```

```python
get_image_feature(s, 5)
```

```python
get_image_feature(s, 50)
```

```python
get_image_feature(s, 500)
```
